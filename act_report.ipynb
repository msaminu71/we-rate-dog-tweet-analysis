## Report: act_report

The twitter account "We Rate Dogs" (@dog_rates) became a twitter sensation in 2016. The handlers of the account repost peoples dogs and rate them based on their unique grading system. Usually, the dogs are rated above 10. This is because they are good dogs.

Following the important wrangling stage of the dataset,involving the gathering of data from various sources, then assessing that data for quality and tidiness issues and finally cleaning the data to make it suitable for analysis. The data can then be explored to provide insights for communication to the stakeholders.

Thorough evaluation of the data showed that majority of the dogs were rated from 10 to 14, however, there was a rating of 1776. I would like to reiterate that the denominator is at least 10. Therefore, there are no bad dogs.

Majority of the dogs were predicted to be "Golden retrievers". However, as no prediction can be totally flawless, some dogs were predicted to be other objects or animals that are not dogs. For example, some dogs were predicted to be "toilet tissue", "paddle", "hammer" amongst others.

The dogtionary showed us that dog usually at various stages at any point in their lives: "doggo", "pupper", "floofer" and "puppo". While most of the tweets were missing the dogs' stage, of the ones with stages provided, "puppo", the transitional phase between "doggo" and "pupper" had the highest frequency.

A visualization of the distribution of the ratings was made. This showed that the most frequent rating was 12, followed by 13, 11, 10 and 14 in that order.
